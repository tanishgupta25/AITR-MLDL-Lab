```python
class Employee:
  def __init__(self,Name,age,salary):
    self.Name=Name;
    self.age=age;
    self.salary=salary;

  def display(self):
      print(f"Employee {self.Name} {self.age} {self.salary}")

  def up(self):
    self.salary=self.salary+100;


empl1 = Employee("Ram","23",12000);
empl1.display();
empl1.up();
empl1.display();
```
