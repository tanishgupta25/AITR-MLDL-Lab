```python
seq = [0,1,2,3,5,8,13]
result = filter(lambda x: x%2 !=0 , seq)
print(list(result))

#result for even numbers
res = filter(lambda x: x%2==0, seq)
print(list(res))
```

```python
List = [[2,3,4],[1,24,16,64],[3,6,8,12]]

#sort each sublist
sortList = lambda x: [sorted(i) for i in x]
print(sortList(List))
```

```python
List = [[2,3,4],[1,24,16,64],[3,6,8,12]]

#sort each sublist
sortList = lambda x: [sorted(i) for i in x]
print(sortList(List))

#second largest
secondLargest = lambda x:[y[-2] for y in sortList(x)]
print(secondLargest(List))
```
