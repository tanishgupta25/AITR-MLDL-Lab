```python
# 2. init function inside class
class Student :
    def __init__(self,name,age) :
        self.name=name
        self.age=age
    def display(self) :
        print("name",self.name)
        print("age",self.age)
s1 = Student("Tanish",22)
s1.display()
```

```python
# 3. instance variable
class Student :
    def __init__(self,name,age) :
        self.name=name   #instance variable
        self.age=age     #instance variable
    def display(self) :
        print("name",self.name)
        print("age",self.age)
s1 = Student("Tanish",22)
s2 = Student("Raj",23)
s1.display()
s2.display()
```

```python
# 4.constructor  & instance variable
class Student :
    def __init__(self,name,age) :  #constructor
        self.name=name   #instance variable
        self.age=age     #instance variable
    def display(self) :
        print("name",self.name)
        print("age",self.age)
s1 = Student("Tanish",22)
s1.display()
```

```python
#5. Polymorphism

# Polymorphism with Functions and Objects - Simple Polymorphism
class dog :
    def speak(self) :
        print("Dog says bow bow")
class cat :
    def speak(self) :
        print("cat says meow meow")
d = dog()
c = cat()
d.speak()
c.speak()
```

```python
# 6.Inheritance
class Parent:
    def __init__(self):
        print("Parent constructor called")

class Child(Parent):
    def __init__(self):
        super().__init__()     # Call parent constructor
        print("Child constructor called")

obj = Child()
```

```python
#7. Multilevel Inheritance

class GrandParent:
    def feature1(self):
        print("Feature 1 from GrandParent")

class Parent(GrandParent):
    def feature2(self):
        print("Feature 2 from Parent")

class Child(Parent):
    def feature3(self):
        print("Feature 3 from Child")

obj = Child()
obj.feature1()
obj.feature2()
obj.feature3()
```

```python
#8. Multiple Inheritance

class Father:
    def skill1(self):
        print("Can drive a car.")

class Mother:
    def skill2(self):
        print("Can cook food.")

class Child(Father, Mother):
    def skill3(self):
        print("Can play football.")

obj = Child()
obj.skill1()
obj.skill2()
obj.skill3()
```

```python
#9. Hierarchical Inheritance

class Animal:
    def sound(self):
        print("Animals make sounds.")

class Dog(Animal):
    def bark(self):
        print("Dog barks.")

class Cat(Animal):
    def meow(self):
        print("Cat meows.")

d = Dog()
c = Cat()

d.sound()
d.bark()

c.sound()
c.meow()
```

```python
#10. Method Overriding - Polymorphism with Inheritance

class Bird:
    def fly(self):
        print("Most birds can fly.")

class Penguin(Bird):
    def fly(self):
        print("Penguins cannot fly.")

# Objects
b1 = Bird()
b2 = Penguin()

b1.fly()   # Calls Bird's method
b2.fly()   # Calls Penguin's method
```

```python
#11. Polymorphism with Functions (Same Function, Different Objects)

class Student:
    def role(self):
        print("I am a student, I study.")

class Teacher:
    def role(self):
        print("I am a teacher, I teach.")

def show_role(person):
    person.role()

s = Student()
t = Teacher()

show_role(s)
show_role(t)
```

```python
#12.Encapsulation

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def show(self):
        print("Name:", self.name)
        print("Marks:", self.marks)

# Create object
s1 = Student("Tanish", 90)
s1.show()
```

```python
#13
class Employee:
    def __init__(self, name, salary):
        self.name = name        # public variable
        self.salary = salary    # public variable

    def display(self):
        print("Name:", self.name)
        print("Salary:", self.salary)

# Create object
e1 = Employee("Tanish", 50000)

# Access public variables and methods directly
print(e1.name)
print(e1.salary)
e1.display()
```

```python
#15
class Student:
    def __init__(self):
        self.name = "Tanish"          # public
        self._grade = "A"            # protected
        self.__marks = 95            # private

    def show(self):
        print("Public Name:", self.name)
        print("Protected Grade:", self._grade)
        print("Private Marks:", self.__marks)

s1 = Student()
s1.show()

print("\nAccessing outside class:")
print("Public:", s1.name)
print("Protected:", s1._grade)
#print("Private:", s1.__marks)   #  Error
```
